```python
import numpy as np
from matplotlib import pyplot as plt
from putils import *
```

```python
def sir_map(S,I,β,γ):
    return [
        S-β*S*I,
        I+β*S*I-γ*I
    ]
```

```python
β = 0.35
γ = 0.1
I = 0.01
R = 0.0
S = 1-I-R
T = 300
Sn = [S]
In = [I]
Rn = [R]
for t in range(T):
    [S,I] = sir_map(S,I,β,γ)
    Sn.append(S)
    In.append(I)
    Rn.append(1-S-I)
plt.plot(Sn,'-');
plt.plot(In,'-');
plt.plot(Rn,'-');
```

```python
def map_osc(x,y,Δt,k):
    return [
        x+Δt*y,
        y-k*Δt*x
    ]
```

```python
Δt = 0.001
k = 40
x = 1
y = 0
xn = [x]
yn = [y]
T = 3000
for t in range(T):
    [x,y] = map_osc(x,y,Δt,k)
    xn.append(x)
    yn.append(y)
plt.plot(xn,yn,'-');
```
